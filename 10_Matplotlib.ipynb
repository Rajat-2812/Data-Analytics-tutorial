```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num = np.arange(5)
num
```

```python
plt.plot(num,num)
plt.plot(num,num*num)
```

```python
# TO make different plot create new figure
plt.plot(num,num)
plt.figure()
plt.plot(num,num*num)
```

```python
plt.figure(figsize=(6,2))
plt.plot(num,num)
plt.figure(figsize=(1,1))
plt.plot(num,num)
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(5,3))
plt.plot(num,num)
plt.plot(num,num*num)
```

```python
plt.bar(num,num*num)
plt.figure()
plt.barh(num,num*num)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.plot(num,num)
plt.title("X")

plt.subplot(1,3,2)
plt.plot(num,num**2)
plt.title("X**2")

plt.subplot(1,3,3)
plt.plot(num,num**3)
plt.title("X**3")
plt.plot()
```
